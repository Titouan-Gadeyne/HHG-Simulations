```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('bmh_TGV1')

import matplotlib.gridspec as gridspec

from mpl_toolkits.axes_grid1.inset_locator import inset_axes

import string

from Optics import Beam, Mask, Offset_beam
```

```python
"CHAPTER 3"
"Q1"

B = Beam(w0=1, lam=500e-9)
M = Mask()

grid = B.Grid(Lx=5e-2, Ly=5e-2, Nx=5000, Ny=5000)

a = 1e-3
L= 1

hx = [-a/2, a/2]
hy = [-a/2, a/2]

S = 0
for x0 in hx:
    for y0 in hy:

        S += M.Iris(*Offset_beam(*grid, z=0, x0=x0, y0=y0)[0:2], R=1e-4)

plt.imshow(S)
plt.xlim(700, 1300)
plt.ylim(700, 1300)
```

```python
FF, grid_FF = B.Fraunhofer(S, grid, 1)

plt.imshow(np.angle(FF), cmap='hsv')
plt.xlim(700, 1300)
plt.ylim(700, 1300)
```

```python
"CHAPTER 3"
"Q2"

B = Beam(w0=1, lam=500e-9)
M = Mask()

grid = B.Grid(Lx=5e-2, Ly=5e-2, Nx=5000, Ny=5000)

a = 1e-3
L= 1

hx = [0, a/2, -a/2]
hy = [np.sqrt(3)*a/3, -np.sqrt(3)*a/6, -np.sqrt(3)*a/6]

S = 0
for (x0, y0) in zip(hx, hy):
        S += M.Iris(*Offset_beam(*grid, z=0, x0=x0, y0=y0)[0:2], R=1e-4)

plt.imshow(S)
plt.xlim(2200, 2800)
plt.ylim(2200, 2800)
```

```python

S2= B.PhaseGrating(*grid, S, np.pi/18)
FF, grid_FF = B.Fraunhofer(S2, grid, 2)

plt.imshow(np.angle(FF), cmap='hsv')
plt.xlim(2200, 2800)
plt.ylim(2200, 2800)
```
